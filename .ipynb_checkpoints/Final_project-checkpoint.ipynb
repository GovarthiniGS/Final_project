Feature Distribution: Visualizing the distribution of continuous features.

```python
continuous_features = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
df[continuous_features].hist(bins=30, figsize=(15, 10))
plt.suptitle('Distribution of Continuous Features')
plt.show()
```

Correlation Matrix: Examine the correlations between features and with the target variable.

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
Visualize relationships between features and the target variable
```

```python
sns.pairplot(df, hue='target', vars=continuous_features)
plt.show()
```

```python
for feature in continuous_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='target', y=feature, data=df)
    plt.title(f'Boxplot of {feature} by Target')
    plt.show()
```

```python
categorical_features = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'NumMajorVessels', 'Thalassemia']
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=feature, hue='target', data=df)
    plt.title(f'Countplot of {feature} by Target')
    plt.show()
```

```python
Chi-square test: Assess the relationship between categorical features and the target variable.
```

```python
from scipy.stats import chi2_contingency

for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df['target'])
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    print(f'Chi-square test for {feature}: chi2 = {chi2:.3f}, p-value = {p:.3f}')
```
